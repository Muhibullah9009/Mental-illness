```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score

df = df.drop(columns=['Entity', 'Code'])

df = df.dropna()
```

```python
X = df.drop(columns=['Depressive disorders (share of population) - Sex: Both - Age: Age-standardized'])
y = df['Depressive disorders (share of population) - Sex: Both - Age: Age-standardized']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(random_state=42),
    'Support Vector Regressor': SVR()
}

results = {}
```

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results[name] = {'MSE': mse, 'R2 Score': r2}

# 7. Show results
for model_name, metrics in results.items():
    print(f"{model_name}:")
    print(f"  MSE: {metrics['MSE']:.6f}")
    print(f"  R2 Score: {metrics['R2 Score']:.4f}")
    print()

# 8. Find best model
best_model = max(results.items(), key=lambda item: item[1]['R2 Score'])
print(f"✅ Best model is: {best_model[0]} with R2 Score: {best_model[1]['R2 Score']:.4f}")
```
